```python
def bananas():
    n = 10
    numtest = 'go'
    numtotal = 0
    numcount = 0
    while numtest == 'go':
        num =input('Enter an integer:')
        if num =='stop':
            numtest ='stop'
            continue
        else:
            numtotal = numtotal +int(num)
            numcount=numcount+1
            n=n-1
    print('Total:',numtotal, 'Count:',numcount,'Average:', numtotal/numcount)
```

```python
bananas()
```
