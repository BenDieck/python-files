```python
#Print keys and associated values using a for loop. 
for key in sorted(stock):
    print(key, stock[key])
```

```python
#Create a list of groceries
groceries = ['apple', 'banana', 'pear']
#Write a function that takes the list as input, and outputs the stock totals
def stocktotal():
    for i in sorted(groceries):
        print(i, stock[i])
```

```python
#Develop a function that will calculate the gross total of the inventory cost. 
#The cooler version would be to throwback an error if there isn't a value, and then continue the code. 
def invgross(): 
    for i in sorted(stock):
        print(i, stock[i]*prices[i])
```
