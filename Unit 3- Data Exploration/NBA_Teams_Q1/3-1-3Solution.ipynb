```python
import os
NBAfile= open('NBA-Attendance-2015.txt','r')
NBAList = []

for line in NBAfile:
    textline=line.strip()
    items = textline.split()
    NBAList.append(items)

len(NBAList)
```

```python
print('Line with Index 0:  ',NBAList[0])


for line in NBAList:
    print('Line: ',line)
```

```python
Avg_attend = []
for line in NBAList:
    Avg_attend.append(int(line[3]))

print(Avg_attend)

```

```python
sum(Avg_attend)

average_att = sum(Avg_attend) / len(Avg_attend)
print(average_att)
```

```python
rank, team, totatt, att, percent = NBAList[0]
print(team)

print(att)

print(percent)

```

```python
percent = []
for line in NBAList:
    percent.append(float(line[4]))
print(percent)
```

```python
percent = [float(line[4]) for line in NBAList]
print(percent)

avg_percent = sum(percent) / len(percent)
print(avg_percent)
```

```python
max_att = max(Avg_attend)
print(max_att)

min_percent = min(percent)
print(min_percent)
```

```python
teams = []
for line in NBAList:
    teams.append(line[1])
print(teams)

print(Avg_attend.index(max_att))


max_team=teams[Avg_attend.index(max_att)]
print('Team with max Average Attendance:  ',max_team)

min_team = teams[percent.index(min_percent)]
print('Team with min Percent of Occupancy:  ',min_team)
```
