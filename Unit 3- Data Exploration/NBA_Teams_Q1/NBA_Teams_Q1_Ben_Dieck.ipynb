```python
NBAfile= open('NBA-Attendance-2015.txt','r')
NBAList = []

for line in NBAfile:
    textline=line.strip()
    items = textline.split()
    NBAList.append(items)

len(NBAList)
```

```python
rank, team, attend, average_attend, perc_occup = NBAList[0]
print(team)

print(average_attend)

print(perc_occup)
```

```python
for line in NBAList:
    print('Line: ',line)
```

```python
team_attend = []
for line in NBAList:
    team_attend.append(int(line[2]))

print(team_attend)
```

```python
avg_g_attend = sum(team_attend)/len(team_attend)
```

```python
max_att = max(team_attend)
print(max_att)

max_percent = max(perc_occup)
print(perc_occup)
```

```python
##Create a list of teams

teams = []
for line in NBAList:
    teams.append(line[1])
print(teams)

```

```python
##Prints the index position in the Avg_attend list where the maxmimum attendence is.
print(team_attend.index(max_att))

```

```python
max_team=teams[team_attend.index(max_att)]
print(max_team)
```

```python
print(perc_occup.index(max_percent))
```

```python
maxp_team=teams[perc_occup.index(max_percent)]
print(maxp_team)
```

```python
print("Sum of Team Attendence:", sum(team_attend))
print("Average attendence:",avg_g_attend)
print("The",max_team,"had the highest attendence of:", max_att)
print("The",maxp_team,"had the highest percentage occupancy of:", perc_occup,"which seems illegal.")
```
