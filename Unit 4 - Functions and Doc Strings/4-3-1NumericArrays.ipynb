```python
import numpy as np          # it is common to make this abbreviation
a = np.array([1, 2, 3])
print(a)

print(type(a))              # python type
print(a.dtype)              # numpy array property data type

```

```python
print(a.shape)              # numpy array property shape

#Contrast this with a two-dimensional array, where a list is given for each row in the array.

b = np.array([[1,2,3],[4,5,6]])
print(b)

print(type(b))

print(b.shape)              # two-dimensional array with two rows and three columns

```

```python
#We use indexing with square brackets, giving an index number from 0 to the length of the dimension—1.  The same indexing conventions apply as in Python, e.g., you can use -1 to get the last element.

print(a[0])

a[0] = 5               # assign a new value to the first location in a
print('Updated one-dimensional array')
print(a)


#For two-dimensional arrays, we use just one square bracket and separate the index numbers by a comma.

print(b[1,2])

b[1,2] = 7
print()
print('Updated two dimensional array')
print(b)
```

```python
#Functions to create arrays

#The numPy package has a number of functions to help create arrays with basic shapes and values.

t = np.zeros((2,2))                  # array of all zeroes
print('Array of all zeroes')
print(t)
print()

t = np.ones((1,2))                  # array of all ones
print('Array of all ones')
print(t)
print()

t = np.full((2,4), 7)                # array of constants (in this case, all 7s)
print('Array of constants')          # this also puts out a future warning
print(t)
print()

t = np.eye(2)                        # identity matrix
print('Identity matrix')
print(t)
print()
#They are used for plotting graphs, statistics and also to do scientific studies and research in almost different fields. 

t = np.random.random((2,2))          # array of random numbers
print('Array of random numbers')
print(t)
print()

print(t.dtype)


#And one final function uses the NumPy function “arange()”, which is analogous to the Python function “range()” to generate a sequence of numbers.

c = np.arange(10)
print(c)
```
