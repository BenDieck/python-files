```python

#Create Lists
examples1 = ['abab', 'xyz', 'aa', 'x', 'bcb']
examples2 = ['', 'x', 'xy', 'xyx', 'xx']
examples3 = ['aaa', 'be', 'abc', 'hello']

# Create a sample that concatinates the three strings

sample = examples1 + examples2 + examples3

# Create a function that iterates over the sample, printing strings of 3 or more while evaluating if the characters at the beginning of the string and the end are the same. 

def output():
    for i in sample:
        if(len(i)>=3 and i[0] == i[-1]):
           print(i)

output()
        
```

```python
#Starting muckrakers
import csv
import numpy as np

#Define a function that combs
def get_scores(source_file):
    SAT_list = []

    with open(source_file, 'r') as csv_file:
        score_reader = csv.reader(csv_file, dialect='excel', delimiter='\t')

        lines = [line for line in score_reader]
        
    for each_line in lines:
        state = str(each_line[0])
        mean_verbal = int(each_line[1])
        mean_math = int(each_line[2])
        SAT_list.append([state, mean_verbal, mean_math])
    return SAT_list

#infile = 'state_satscores_2004.tsv'
```

```python

# Initialize with negative infinity to ensure any value in the list will be greater - This was a gitHUB suggestion.
max_mean_verb = float('-inf')  

state_with_max_mean_verb = ""

#This subsequently evaluates whether each state is higher than the one before it. I feel like this is different than indexing to the max value, which I feel like we've done? targ
for i in SAT_list:
    state, mean_verbal, mean_math = i
    if mean_verbal > max_mean_verb:
        max_mean_verb = mean_verbal
        state_with_max_mean_verb = state

print("State with the largest Mean Verbal Score:", state_with_max_mean_verb)
print("High Verbal Mean:", max_mean_verb)
```
