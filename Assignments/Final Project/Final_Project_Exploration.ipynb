```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
```

```python
def get_song_info(source_file):
    song_list=[]
    with open(source_file,'r') as csv_file:
        song_reader=csv.reader(csv_file,dialect='excel',delimiter='\t')
        lines=[line for line in song_reader]
    try:    
        for each_line in lines:
                if line[0] == '' or line[0].startswith('track_') or line[0].startswith('#') or line[0].startswith('playlist_') or line[0].startswith('songs_') or line[0].startswith('danceability') or line[0].startswith('energy') or line[0].startswith('key') or line[0].startswith('loudness') or line[0].startswith('mode') or line[0].startswith('speechiness') or line[0].startswith('acousticness') or line[0].startswith('instrumentalness') or line[0].startswith('liveness') or line[0].startswith('valence') or line[0].startswith('tempo') or line[0].startswith('duration_'):
                  continue
                else:
                        song_id=str(each_line[0])
                        count_song_id=int(each_line[1])
                        song_name=str(each_line[2])
                        artist=str(each_line[3])
                        popularity=int(each_line[4])
                        album_id=str(each_line[5])
                        album_name=str(each_line[6])
                        album_release_date=str(each_line[7])
                        album_year=str(each_line[8])
                        playlist_count=int(each_line[9])
                        playlist_id=str(each_line[10])
                        genre=str(each_line[11])
                        subgenre=str(each_line[12])
                        dance_score=float(each_line[13])
                        energy=float(each_line[14])
                        key=int(each_line[15])
                        loudness=float(each_line[16])
                        mode=int(each_line[17])
                        spe_score=float(each_line[18])
                        acou_score=float(each_line[19])
                        instru_score=float(each_line[20])
                        live_score=float(each_line[21])
                        valence=float(each_line[22])
                        tempo=float(each_line[23])
                        duration=float(each_line[24])
                        song_list.append([song_name, artist, genre, sub_genre, album_name, 
                                        album_release_date, album_year, popularity, dance_score,
                                        energy, key, loudness, spe_score, acou_score, instru_score,
                                        live_score, valence, tempo, duration])
    except UnicodeDecodeError:
        print ("Error occurred", line)
    return song_list
                

```
