```python
   #Create this as a function
    
    def evaluate_numbers(numbers):
    #Creates dictionary, occurrences
    occurrences = {}
    
    #Validate the input so that it only takes numeric input.
    for number in numbers:
        if not isinstance(number, (int, float)):
            raise ValueError("Non-numeric value found in the input list.")
    
    # Count occurrences of each unique number
    for number in numbers:
        if number in occurrences:
            occurrences[number] += 1
        else:
            occurrences[number] = 1
        
    # Check conditions, get the 0 slotted value for the key 19 or the key 5.
    
    count_19 = occurrences.get(19, 0)
    count_5 = occurrences.get(5, 0)
    
    # Evaluate the expression
    
    if count_19 >= 2 and count_5 >= 3:
        result_message = "Conditions met: At least two numbers are 19, and at least three numbers are 5.\n"
        # Plus equal adds another line
        result_message += "Number Counts:"
        # For the key which we are calling number, and the thing we are indexing, which we are calling count, show the count. 
        for number, count in occurrences.items():
            result_message += f"\n{number}: {count}"
        return result_message
    
    # If the above conditions aren't meant, send this. 
    else:
        return "Conditions not met."
```

```python
# Example 1
my_numbers = [19,19,15,5,3,5,5,2]

result = evaluate_numbers(my_numbers)
#Gives the formatted print statement
print(result)
```

```python
# Example 2

my_numbers = [19,15,15,5,3,3,5,2]

result = evaluate_numbers(my_numbers)
#Gives the formatted print statement
print(result)
```

```python
#Define the new function

def string_lengths(strings):
    
     #Validate the input so that all input values must be a string
        
    if not all(isinstance(s, str) for s in strings):
        raise TypeError("All elements in the input list must be strings.")
    
    #Print a heading for the output
    print("List of strings:")
    
    #Create variables to hold evaluated strings
    lengths = {}
    longest_string = ""
    longest_length = 0
    
    # Print out each string and the length of the string
    for s in strings:
        print(f"String: {s}, Length: {len(s)}")
    
    # Evaluate each string to see if its longer than the last one... If it is, replace my variable. 
    for s in strings:
        current_length = len(s)
        lengths[s] = current_length

        if current_length > longest_length:
            longest_string = s
            longest_length = current_length

    print(f"Longest String in Collection: {longest_string}, Length: {longest_length}")
    
```

```python
list1 = ['cat', 'car', 'fear', 'center']

string_lengths(list1)
```

```python
list2 = ['cat', 'dog', 'shatter', 'donut', 'at', 'todo', '']

string_lengths(list2)
```
