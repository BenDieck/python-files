```python
word_freq = {'the':58, 'people':6, 'beautiful':8, 'cats':13, 'finally':9, 'this': 19, 'class': 21, 'is': 101, 'almost': 1, 'over':4}
```

```python
def eval_freq(i_dict,t):
    above_threshold = {}
    for key, value in i_dict.items():
        if value > t:
            above_threshold[key] = value
    
    return above_threshold
    
```

```python
eval_freq(word_freq, 10)
```

```python
import pandas as pd
#Read in the data
df = pd.read_csv('eur.pop.rev.csv', delimiter=',', header = 0)
# Rename the screwy column names
df = df.rename(columns={'European Population (1989-1995)':'Country', 'Unnamed: 1':'1989', 'Unnamed: 2':'1990', 'Unnamed: 3':'1991', 'Unnamed: 4':'1992',
                        'Unnamed: 5':'1993', 'Unnamed: 6':'1994', 'Unnamed: 7':'1995'})
# Drop the whitespace
df = df.drop([0, 1])
# Replace NA data with 0
df = df.fillna(0)
# Change columns to numeric
df['1989'] = pd.to_numeric(df['1989'])
df['1990'] = pd.to_numeric(df['1990'])
df['1991'] = pd.to_numeric(df['1991'])
df['1992'] = pd.to_numeric(df['1992'])
df['1993'] = pd.to_numeric(df['1993'])
df['1994'] = pd.to_numeric(df['1994'])
df['1995'] = pd.to_numeric(df['1995'])

print('Countries over a million avg population, 1995')
print('')
for index, row in df.iterrows():
    if row['1995'] > 1000000:
        print(f"Country: {row['Country']}, Population '95: {row['1995']}")
```
