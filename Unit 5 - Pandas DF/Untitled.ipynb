```python
## Load muckrakers
import pandas as pd
```

```python
## Load information into a usable  df

persondict = {'person':['Bob','Alice','Steve'],'age':[32,24,64],'weight':[128,86,95]}
my_df=pd.DataFrame(persondict, columns=['person','age','weight'])
my_df
```

```python
## Reset the index to be the person column 
my_df = my_df.set_index('person')
my_df
```

```python
## Stack the df
result = my_df.stack()
result
```

```python
## Reset the index
tall_df = result.reset_index()
tall_df
```

```python
## Rename the columns

tall_df.columns=['person','attribute','value']
tall_df
```

```python
## Convert back to the original shape
result.unstack()
```
