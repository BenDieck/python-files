```python
# pip install fredapi
from fredapi import Fred
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_data(api_key):

    fred = Fred(api_key = api_key)
    data = fred.get_series('SP500')
    df = pd.DataFrame({'date':data.index, 'price': data.values})
    return df
```

```python
def graph_data(data):
    rollup_by_year = data.groupby(data.date.dt.year)['price'].sum()
    graph = rollup_by_year.plot.bar()
    plt.show()
```

```python
def main():
    api_key = os.environ["FRED_API_KEY"]
    data = get_data(api_key)
    graph_data(data)
```

```python
main()
```
