```python
#!pip install pymongo
import os
import json
import pymongo
import pandas as pd
```

```python
def find_movies(database_name, collection_name, host_name, host_port, movie_name):  
    try:
        client = pymongo.MongoClient(host_name, host_port)
        print ("Connected successfully!!!")
    except e:
        print ("Could not connect to MongoDB: %s" % e )
    else:

        # use database named usgs or create it if not there already
        movie_db = client[database_name]
            # create collection named earthquakes or create it if not there already
        movie_coll = movie_db[collection_name]
        
        movie_list = {}
        for doc in movie_coll.find({"name": movie_name}):
            show_name = doc['name']
            show_type = doc['type']
            language = doc['language']
            url = doc['url']
            officialSite = doc['officialSite']
            image_url = doc['image']['medium']
            movie_list[show_name] = ([show_name, show_type, language, url, officialSite, image_url])
        client.close()
        
        for show_name in movie_list:
            print(show_name)
        
        df = pd.DataFrame(movie_list.values(), columns=['name', 'type', 'language', 'url', 'officialSite', 'image_url'])
        return df
```

```python
folder_name = '../data/netflix-canceled-2021'
database_name = 'neflix2021'
collection_name = 'movies'
host_name = 'localhost'
host_port = 27017

```

```python
movie_name = "Grand Army"
movie_list = find_movies(database_name, collection_name, host_name, host_port, movie_name)
movie_list
```
