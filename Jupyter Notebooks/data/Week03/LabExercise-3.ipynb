Consider the following two dictionaries:

Show the expression that gets the value of the stock dictionary at the key ‘orange’. Print out the results.
```python
print(f"The value of the stock dictionary at the key 'orange' is '{stock['orange']}'")
stock['cherry'] = 50
prices['cherry'] = 2.7

print('The stock dictionary')
print(stock)

print('The prices dictionary')
print(prices)
```

Write the code for a loop that iterates over the stock dictionary and prints each key and value.

```python
print("print the stock list using for loop")
for k, v in stock.items():
    print(f"Item '{k}' has {v} items in stock")
```

c. Suppose that we have a list:

```python
groceries = ['apple', 'banana', 'pear']
print(groceries)
```

Write the code that will sum the total number in stock of the items in the groceries list. How many apples + bananas + pears are there? Display the total items labeled appropriately.

```python
total_in_stock = 0
for item in groceries:
    total_in_stock += stock[item]
print(f"Total groceries in stock using a for loop: {total_in_stock}")
total_in_stock = sum(map(lambda item: stock[item], groceries))
print(f"Total groceries in stock using lambda: {total_in_stock}")
```

d. Write the code that can print out the total value in stock of all the items. This program can iterate over the stock dictionary and for each item multiply the number in stock times the price of that item in the prices dictionary. (This can include the items for ‘cherry’ or not, as you choose.) Display the total value with appropriate label and formatting of the total value in dollars.

```python
total_value = 0.0
for item in stock:
    total_value += float(stock[item]) * prices[item]
print(f"Total value of prices and stock using for loop: {total_value}")

total_value = sum(map(lambda item: prices[item] * float(stock[item]), stock))
print(f"Total value of prices and stock using lambda: {total_value}")
total_value =sum(prices[item] * stock[item] for item in prices)
print(f"Total value of prices and stock using sum: {total_value}")
```
