```python
import pymongo
from pymongo import MongoClient
```

```python
client = MongoClient('localhost', 27017)
```

```python
# show existing databases
client.list_database_names()
```

```python
# try making a db
db = client.sampledb
peoplecoll = db.people
```

```python
# list available collections within the db
db.list_collection_names()
```

```python
# see if it worked
client.list_database_names()
```

```python
peoplelist = [{ "name": "Britney Sykes", "age": 21 , 'position':'Guard'}, { "name": "Briana Day", "age": 19, 'position':'Center'}, 
              { "name": "Alexis Peterson", "age": 21, 'position':'Guard' }, { "name": "Gabby Cooper", "age": 18, 'position':'Guard'}]
```

```python
 peoplecoll.insert_many(peoplelist[0: ])
```

```python
# see if it worked
client.list_database_names()
```

```python
# see if the collection is there
db.list_collection_names()
```

```python
docs = peoplecoll.find()
for doc in docs:
    print(doc)
```

```python
# find all the documents where the position field has an age less than 20
# I had to google what query operators were, that's the bananas dollar sign looking thing. 
results = peoplecoll.find({"age": {"$lt": 20}})
for result in results:
    print(result)
```
