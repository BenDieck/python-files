```python
# try accessing a valid API server using an invalid URL
api_url = "http://api.open-notify.org/this-api-doesnt-exist"
response = requests.get(api_url)
```

```python
# display the results
print(response.status_code)

if response.status_code == 200:
    print(response.json())
```
