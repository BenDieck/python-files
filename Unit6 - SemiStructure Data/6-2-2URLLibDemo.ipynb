```python
from urllib import request
snapchaturl = "http://www.bbc.com/news/business-39135278"
## or altogether html =
request.urlopen(snapchaturl).read().decode('utf8')
response = request.urlopen(snapchaturl)
type(response)
```

```python
html = response.read().decode('utf8')
type(html)
```

```python
print(html[:500])
```

```python
from bs4 import BeautifulSoup
htmlsoup = BeautifulSoup(html, 'html.parser')
type(htmlsoup)
```

```python
firsttitle = htmlsoup.title
print(firsttitle)
print(type(firsttitle))
print(firsttitle.name)
```

```python
firsttitle.get_text()
```

```python
anchors = htmlsoup.find_all('a')
print(type(anchors))
print(len(anchors))
```

```python
print(anchors[:8])
```

```python
links = [str(link.get('href')) for link in htmlsoup.find_all('a')]
print(links[ :4])
```

```python
outlinks = [link for link in links if link.startswith('http')]
print(len(outlinks))

print(outlinks[:4])
```
