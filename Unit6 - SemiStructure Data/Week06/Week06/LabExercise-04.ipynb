```python
from functools import reduce
```

```python
def mult_loop(list):
    prod = 1
    for item in list:
        prod *= item # prod = prod * item
        print(item, prod)
    return prod
```

```python
def mult_lambda(list):
    return reduce(lambda x, y: x*y, list)
```

Write a Python function that accepts a string and calculates the number of upper case letters 
and lower case letters. 
Sample String : 'The quick Brown Fox’ 
Execute the function with at least 3 different phrases. 

```python
def count_case_loop(str):
    count_upper = 0
    count_lower = 0
    for char in str:
        print(char)
        if char.isupper():
            count_upper += 1
        elif char.islower():
            count_lower += 1
        else:
            pass
    return count_upper, count_lower
```

```python
def count_case_sum(str):
    count_upper = sum(1 if char.isupper() else 0 for char in str)
    count_lower = sum(1 if char.islower() else 0 for char in str)
    return count_upper, count_lower
```

```python
count_upper, count_lower = count_case_loop("The quick Brown Fox")
print(f"upper: {count_upper}, lower: {count_lower} (loop)")
```

```python
count_upper, count_lower = count_case_sum("The quick Brown Fox")
print(f"upper: {count_upper}, lower: {count_lower} (sum)")
```
